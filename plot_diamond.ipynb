```python
import matplotlib.pyplot as plt
import numpy as np
Ry2eV = 13.6
```

```python
detot = np.array([1.26944,
1.27931,
1.28217,
1.27873,
1.27579])*Ry2eV

x=np.array([1,2,3,4,5])

fig,ax = plt.subplots(figsize = (8,6), dpi = 100)
#ax.plot(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1], label='Gamma only')
#ax.plot(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1], label = '3x3 w/o spin')
ax.plot(1/x**3,detot, label='3x3 with spin')
#ax.scatter(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1])
#ax.scatter(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1])
ax.scatter(1/x**3,detot)

plt.ylabel('eV')
plt.xlabel('1/Volume')
plt.title('Total Energy differences')
plt.legend()
#plt.xlim(0,0.03)
plt.ylim(17, 18)
plt.show()
```

```python
dehart = np.array([0.183517,
0.115209,
0.12833,
0.165711,
0.170775])*Ry2eV

x=np.array([1,2,3,4,5])

fig,ax = plt.subplots(figsize = (8,6), dpi = 100)
#ax.plot(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1], label='Gamma only')
#ax.plot(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1], label = '3x3 w/o spin')
ax.plot(1/x**3,dehart, label='3x3 with spin')
#ax.scatter(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1])
#ax.scatter(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1])
ax.scatter(1/x**3,dehart)

plt.ylabel('eV')
plt.xlabel('1/Volume')
plt.title('Hartree Energy differences')
plt.legend()
#plt.xlim(0,0.03)
plt.ylim(1.5, 2.6)
plt.show()
```
