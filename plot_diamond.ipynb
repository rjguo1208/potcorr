```python
import matplotlib.pyplot as plt
import numpy as np
Ry2eV = 13.6
```

```python
detot = np.array([1.26944,
1.27931,
1.28217,
1.27873,
1.27579])*Ry2eV

x=np.array([1,2,3,4,5])

fig,ax = plt.subplots(figsize = (8,6), dpi = 100)
#ax.plot(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1], label='Gamma only')
#ax.plot(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1], label = '3x3 w/o spin')
ax.plot(1/x**3,-detot, label='3x3 with spin')
#ax.scatter(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1])
#ax.scatter(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1])
ax.scatter(1/x**3,-detot)

plt.ylabel('eV')
plt.xlabel('1/Volume')
plt.title('Total Energy differences')
plt.legend()
#plt.xlim(0,0.03)
plt.ylim(-18, -17)
plt.show()
```

```python
detot_lec = np.array([-0.1,
0.04,
0.14])
detot_cal = -1*np.array([1.27931,
1.28217,
1.27873,])*Ry2eV

x=np.array([2,3,4])
fig,ax = plt.subplots(figsize = (8,6), dpi = 100)
#ax.plot(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1], label='Gamma only')
#ax.plot(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1], label = '3x3 w/o spin')
ax.plot(1/x**3,detot_cal-detot_cal[-1], label='calculated')
ax.plot(1/x**3,detot_lec-detot_lec[-1], label='Pasquarello')
#ax.scatter(1/dtote_gamma_x[::-1]**2,dtote_gamma[::-1])
#ax.scatter(1/dtote_3x3_x[::-1]**2,dtote_3x3[::-1])
ax.scatter(1/x**3,detot_cal-detot_cal[-1])
ax.scatter(1/x**3,detot_lec-detot_lec[-1])

plt.ylabel('eV')
plt.xlabel('1/Volume')
plt.title('Total Energy differences')
plt.legend()
#plt.xlim(0,0.03)
#plt.ylim(17, 18)
plt.show()
```
